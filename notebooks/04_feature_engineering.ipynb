```python
import numpy as np
import pandas as pd

df = pd.read_csv('../data/processed/flights_day3.csv')
print(df.shape)
```

```python
df['IS_WEEKEND'] = df['DAY_OF_WEEK'].apply(lambda x: 1 if x in [6, 7] else 0)
print(df['IS_WEEKEND'].value_counts())
```

```python
df['ROUTE'] = df['ORIGIN_AIRPORT'] + '_' + df['DESTINATION_AIRPORT']
route_counts = df['ROUTE'].value_counts()
df['ROUTE_POPULARITY'] = df['ROUTE'].map(route_counts)

print(df[['ORIGIN_AIRPORT', 'DESTINATION_AIRPORT', 'ROUTE', 'ROUTE_POPULARITY']].head())
print(f"\nUnique routes: {df['ROUTE'].nunique()}")
```

```python
airline_reliability = df.groupby('AIRLINE')['IS_DELAY'].mean()
df['AIRLINE_DELAY_RATE'] = df['AIRLINE'].map(airline_reliability)

print(df[['AIRLINE', 'AIRLINE_DELAY_RATE']].drop_duplicates().sort_values('AIRLINE_DELAY_RATE', ascending=False))
```

```python
df_day4 = df.drop(columns=['AIRLINE_DELAY_RATE'])


df_day4.to_csv('../data/processed/flights_day4.csv', index=False)
print("Saved!", df_day4.shape)
```
