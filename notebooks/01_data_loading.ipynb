```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv('../data/raw/flights.csv', low_memory=False)
print(df.shape)
print(df.dtypes)
```

```python
df.isnull().sum().sort_values(ascending=False).head(15)
```

```python
print(f"Before drop: {df.shape}")
df = df.dropna(subset=['ARRIVAL_DELAY']).copy()
print(f"After drop: {df.shape}")
print(f"Dropped: {105071} rows ({105071/5819079*100:.2f}%)")
```

```python
df['IS_DELAY'] = np.where(df['ARRIVAL_DELAY'] > 15, 1, 0)
print(df['IS_DELAY'].value_counts())
print(df['IS_DELAY'].value_counts(normalize=True)*100)
```

```python
df.to_csv('../data/processed/flights_cleaned.csv', index=False)
print("Saved!")
```
